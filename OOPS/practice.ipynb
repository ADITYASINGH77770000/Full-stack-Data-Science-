1. Write a class Animal with a method sound().
Create a child class Dog that overrides the method.

```python
class Animal:
    def speak(self):
        return "sound of animal"
    
class Dog(Animal):
    def speak(self):
        return "woof!"

class Cat(Animal):
    def speak(self):
        return "meow!"
    
def get_animal(Animal):
    print(Animal.speak())

dog1=Dog()
cat1=Cat()
get_animal(dog1)
get_animal(cat1)
```

2. Create a class Student that stores marks of 3 subjects.
Add a method to calculate total and average marks.

```python
class Student:
    def __init__(self,name,subject1,subject2,subject3):
        self.name=name
        self.subject1=subject1
        self.subject2=subject2
        self.subject3=subject3
    
    def average(self):
        return (self.subject1 + self.subject2 + self.subject3) / 3
    
    def sum(self):
        return self.subject1 + self.subject2 + self.subject3
    
student1=Student("Alice",85,90,95)
print("Average marks:", student1.average())     
print("Total marks:", student1.sum())
```

3. Create a class Movie with attributes name, rating.
Add a method to display movies with rating above 8.

```python
class Movie():
    def __init__(self,name,rating):
        self.name=name
        self.rating=rating
        
    def rating_category(self):
        if self.rating >= 8:
            return "Excellent"
       
movie1=Movie("Inception",9)
movie2=Movie("tent",7)
print("Movie Rating Category:", movie1.rating_category())
print("Movie Rating Category:", movie2.rating_category())
```

4. Create a class OnlineCourse with methods to enroll students and display enrolled count.

```python
class course():
    def __init__(self,course_name,course_id,student_name):
        self.course_name=course_name
        self.course_id=course_id
        self.student_name=student_name
    
    def enroll_student(self):
        return f"Student {self.student_name} has been enrolled in {self.course_name} (ID: {self.course_id})"
    
    def enroll_count(self, count):
        return f"Total enrolled students: {count}"

course1=course("Mathematics","MATH101","Bob")
print(course1.enroll_student())
print(course1.enroll_count(30))
```

```python
class OnlineCourse:
    def __init__(self, course_name):
        self.course_name = course_name
        self.students = []

    def enroll_student(self, student_name):
        self.students.append(student_name)
        print(f"{student_name} has been enrolled in {self.course_name}.")

    def display_enrolled_count(self):
        print(f"Total students enrolled in {self.course_name}: {len(self.students)}")


# Example usage
course = OnlineCourse("Python for Data Science")

course.enroll_student("Aditya")
course.enroll_student("Rahul")
course.enroll_student("Sneha")

course.display_enrolled_count()
```

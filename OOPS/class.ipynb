```python
## Instance Variable and methods
class Dog:
    ## Constructor
    def __init__(self,name,age,no,color):
        self.name=name
        self.age=age
        self.no=no
        self.color=color

## Create Objects 
dog1=Dog("Buddy",3,12,"Red")
print(dog1)
print(dog1.name)
print(dog1.age)
print(dog1.color)
```

```python
## Define a class with instance methods

class Dog:
    ## Constructor
    def __init__(self,name,age,no,color):
        self.name=name
        self.age=age
        self.no=no
        self.color=color
        
    def bark(self):
        print(f"{self.name} says woof")
        
dog1=Dog("Buddy",3,12,"Red")
dog1.bark()
dog2=Dog("Lucy",5,8,"Blck")
dog2.bark()
```

```python
## Modeling a bank account 

class BankAccount:
    def __init__(self,Name,owner,balance=0):
        self.Name=Name
        self.owner=owner
        self.balance=balance
        
    def deposit(self,amount):
        self.balance+=amount
        print(f"Deposited {amount}. New balance is {self.balance}")
    
    def withdraw(self,amount):
        if amount>self.balance:
            print("Insufficient funds")
        else:
            self.balance-=amount
            print(f"Withdrew {amount}. New balance is {self.balance}")
    
    def get_balance(self):
        return self.balance
    
## create an account 

account=BankAccount("HDFC","Alice",1000)
print(account.balance)
```

```python
## call instance methods
account.deposit(100)
```

```python
account.withdraw(500)
```

```python
print(account.get_balance())
```

2. Create a class Employee and:

calculate annual salary

give bonus using a method

```python
class Employee:
    def __init__(self,name,position,salary=0):
        self.name=name
        self.position=position
        self.salary=salary
    
    def annual_salary(self):
        return self.salary*12
    
    def bonus(self,percentage):
        return self.salary*(percentage/100)
    
    def get_salary(self):
        return self.salary
```

```python
employee1=Employee("John Doe","Manager",5000)
employee2=Employee("Jane Smith","Developer",4000)
print(employee1.name)
print(employee1.annual_salary())
print(employee2.bonus(10))
print(employee2.get_salary())
```

3. Write a class Rectangle with methods to calculate:

area

perimeter

```python
class rectangle:
    def __init__(self,length,breadth):
        self.length=length
        self.breadth=breadth
        
    def area(self):
        return self.length*self.breadth
    
    def perimeter(self):
        return 2*(self.length+self.breadth)
    
    def get_length(self):
        return self.length
    
```

```python
rect1=rectangle(10,5)
print(rect1.length)
print(rect1.area())
print(rect1.perimeter())
print(rect1.get_length())
```

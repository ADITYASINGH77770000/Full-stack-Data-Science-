1. Count vowels, consonants, digits, and spaces in a text file

It should return:

number of vowels

number of consonants

number of digits

number of spaces

```python
def count_words(file_path):
    with open(file_path, 'r') as file:
        lines = file.read()
        vowels = 'aeiouAEIOU'
        vowels_count = 0
        consonants_count = 0
        digits_counts = 0
        spaces_count = 0
        
        for char in lines:
            if char in vowels:
                vowels_count += 1
            elif char.isalpha():
                consonants_count += 1
            elif char.isdigit():
                digits_counts += 1
            elif char.isspace():
                spaces_count += 1
    return vowels_count, consonants_count, digits_counts, spaces_count
```

```python
file_path = 'new.txt'
vowels,consonants,digits,spaces = count_words(file_path)
print(f"vowles: {vowels}, consonants: {consonants}, digits: {digits}, spaces: {spaces}")
```

## File Operation - Read and write Files

```python
### Read a whole fle 
with open('example.txt','r') as file:
     content=file.read()
     print(content)
```

```python
## Read a file line by line
with open('example.txt','r') as file:
    for line in file:
        print(line.strip())  # strip() to remove leading/trailing whitespace
```

```python
## writing a file (overwriting)
with open('example.txt','w') as file:
    file.write("This is a new line.\n")
    file.write("This will overwrite existing content.\n")
```

```python
## write a file (without overwriting)
with open('example.txt','a') as file:
    file.write("This line is appended.\n")
```

```python
## writing a list of lines to a file 
lines = ["First line\n", "Second line\n", "Third line\n"]
with open('example.txt','a') as file:
    file.writelines(lines)
```

```python
## Binary files
# writing binary file
data = b'\x00\x01\x02\x03\x04'
with open('example.bin', 'wb') as file:
    file.write(data)
```

```python
### reading a binary file 
with open('example.bin','rb') as file:
    content = file.read()
    print(content)  # Output will be in bytes
```

```python
### Read the content from a source text file and write it to a destination text file 
# copying a text file 
with open('example.txt','r') as source_file:
    content = source_file.read()
with open('copy_example.txt','w') as destination_file:
    destination_file.write(content)
```

```python
## read a text file and count the number of words , lines and characters
# counting lines, words, and characters in a text file
def count_text_file(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
        line_count = len(lines)
        word_count = sum(len(line.split()) for line in lines)
        char_count = sum(len(line) for line in lines)
    return line_count, word_count, char_count

file_path = 'example.txt'
lines,words,characters = count_text_file(file_path)
print(f"lines: {lines},words: {words}, Characters: {characters}")
        
```

```python
### writig and then recording a file
with open('example.txt','w+') as file:
    file.write("This is a new line. \n")
    file.write("This will overwrite existing content. \n")
    
    ## Move the file cursor to the begining 
    file.seek(0)
    
    ## read the content of the file 
    content = file.read()
    print(content)
```

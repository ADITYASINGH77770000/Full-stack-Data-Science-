### Conditional Statements (if, elif, else) 

```python
## if statement
age=20

if age>=18:
    print("You are an adult.")
```

```python
## else statement
## The else statement is used to execute a block of code when the condition in the if statement is false.

age = int(input("Enter a number:"))
if age>=18:
    print("You are an eligible for voting.")
else:
    print("You are a minor.")
```

```python
## elif 
## The elif statement is used to check multiple conditions in a sequence.
age = int(input("Enter your age: "))
if age < 14:
    print("You are a minor.")
elif age < 18:
    print("You are a teenager.")
else:
    print("You are an adult.")   
```

```python
## Nested conditional statements

## You can nest if statements inside other if statements to create more complex conditions.

num = int(input("Enter your num: "))

if num>0:
    print("You entered a positive number.")
    if num%2==0:
        print("The number is even.")
    else:
        print("The number is odd.")
    
else:
    print("The number is negative or zero")
```

```python
# Practical Examples

## Determine if a year is a leap year using nested condition statement
year = int(input("Enter a year: "))

if year %4 == 0:
    if year % 100 == 0:
        if year % 400 == 0:
            print(f"{year} is a leap year.")
        else:
            print(f"{year} is not a leap year.")
    else:
        print(f"{year} is a leap year.")
else:
    print(f"{year} is not a leap year.")
```

```python
## Assignment 
## sample calculator programame
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
operation = input("Enter operation (+, -, *, /): ")

# Perform the operation based on user input
if operation == '+':
    result = num1 + num2
elif operation == '-':
    result = num1 - num2
elif operation == '*':
    result = num1 * num2
elif operation == '/':
    if num2 != 0:
        result = num1 / num2
    else:
        result = "Error: Division by zero is not allowed."
else:
    result = "Error: Invalid operation."
print(f"The result is: {result}")
```

```python
## Employee bonus calculation

# Take user input for employee's performance rating

years_of_service = int(input("Enter the number of years worked: "))
performance_rating = float(input("Enter your performance rating (0-5): "))

# Determine bonus percentage 
if performance_rating >= 4.5:
    if years_of_service > 10:
        bonus_percentage = 20
    elif years_of_service > 5:
        bonus_percentage = 15
    else:
        bonus_percentage = 10
elif performance_rating >= 3.0: 
    if years_of_service > 10:
        bonus_percentage = 10
    elif years_of_service > 5:
        bonus_percentage = 5
    else:
        bonus_percentage = 2
else:
    bonus_percentage = 0
   
# calculated bonus amount 
salary = float(input("Enter your salary: "))
bonus_amount = salary*bonus_percentage/100
print("Bonus Amount: $(:.2f)".format(bonus_amount)) 
```

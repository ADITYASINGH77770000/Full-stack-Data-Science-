## Map Function

```python
def square(x):
    return x*x

square(10)
```

```python
# map function without lambda 

numbers=[1,2,3,4,5,6,7,8]

list(map(square,numbers))
```

```python
## lambda function with map
numbers=[1,2,3,4,5,6,7,8]
list(map(lambda x:x*x,numbers))
```

```python
### Map multiple iterabales 

numbers1=[1,2,3]
numbers2=[4,5,6]

added_numbers=list(map(lambda x,y:x+y,numbers1,numbers2))
print(added_numbers)
```

```python
words=['hello','world','python']
upper_word=list(map(str.upper,words))
print(upper_word)
```

```python
def get_name(person):
    return person['name']

people=[
    {'name': 'Alice', 'age': 30},
    {'name': 'Bob', 'age': 25},
    {'name': 'Charlie', 'age': 35}
]
```

```python
list(map(get_name, people))
```

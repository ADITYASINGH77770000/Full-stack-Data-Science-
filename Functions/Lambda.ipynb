## Lambda 

```python
# syntax

lambda arguments: expression
```

```python
def addition(a,b):
    return a + b
```

```python
addition(2,3)
```

```python
addition=lambda a,b: a + b
type(addition)
print(addition(2,3))
```

```python
def even(num):
    if num % 2 == 0:
        return True
    
even(24)
             
```

```python
even1=lambda num: num % 2 == 0
even1(24)
```

```python
def addition(x,y,z):
    return x + y + z

addition(12,13,13)
```

```python
additional=lambda x,y,z: x + y + z
additional(12,13,13)
```

```python
## map() - applies a function to all the items in an input list 

numbers=[1,2,3,4,5,6]
def square(num):
    return num ** 2

square(2)
```

```python
list(map(lambda x:x**2,numbers))
```

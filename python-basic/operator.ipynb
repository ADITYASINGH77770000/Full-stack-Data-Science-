## Operators

```python
## Arithmetic Operators

a=10
b=5

add_result = a + b
sub_result = a - b
mul_result = a * b
div_result = a / b
mod_result = a % b
exp_result = a ** b
floor_div_result = a // b
modulus_result = a % b

print("Addition:", add_result)
print("Subtraction:", sub_result)
print("Multiplication:", mul_result)
print("Division:", div_result)
print("Modulus:", mod_result)
print("Exponentiation:", exp_result)
print("Floor Division:", floor_div_result)
print("Modulus:", modulus_result)
```

# Comparison Operators

```python
a=10
b=10
print(a == b)  # True, because a and b are equal
```

```python
str1 = "Hello"
str2 = "World"
result = str1 + " " + str2
print(result)  # This will print "Hello World"
```

```python
## Not equal to !=
str1!= str2
```

```python
# greater than >
num1=45
num2=30
print(num1 > num2)  # True, because 45 is greater than 30
```

```python
## less than <

print(num1<num2)
```

```python
## Greater than or equal to >=
print(num1 >= num2)  # True, because 45 is greater than or equal to 30
```

```python
## less than or equal to <=
print(num1 <= num2)  # False, because 45 is not less than or
```

## Logical Operators

```python
## And ,Not,or
X=True
Y=True
result =X and Y 
print(result)
```

```python

X=False
Y=True

result =X and Y
print(result) 
```

```python
 ## or
x = True
y = False
result = x or y
print(result)  # True, because at least one operand is True
```

```python
# Not operator
x = True
result = not x
print(result)  # False, because not True is False
```

```python
# Simple Calculator

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

# Perform arithmetic operations
addition = num1 + num2
subtraction = num1 - num2
multiplication = num1 * num2
division = num1 / num2 if num2 != 0 else "undefined (division by zero)"
floor_division = num1 // num2 if num2 != 0 else "undefined (division by zero)"
modulus = num1 % num2 if num2 != 0 else "undefined (division by zero)"
exponentiation = num1 ** num2

# Display results
print(f"Addition: {num1} + {num2} = {addition}")
print(f"Subtraction: {num1} - {num2} = {subtraction}")
print(f"Multiplication: {num1} * {num2} = {multiplication}")
print(f"Division: {num1} / {num2} = {division}")
print(f"Floor Division: {num1} // {num2} = {floor_division}")
print(f"Modulus: {num1} % {num2} = {modulus}")
print(f"Exponentiation: {num1} ** {num2} = {exponentiation}")
```

### Dictonary


```python
## Creating Dictionaries

empty_dict={}
print(type(empty_dict))
```

```python
empty_dict=dict()
print(type(empty_dict))
```

```python
student={"name":"John", "age":20, "course":"Python"}
print(student)
print(type(student))
```

```python
# Single key is always used
student={"name":"John", "age":20, "name":"Python"}
print(student)
```

```python
# accessing Dictionary elements
student={"name":"john", "age":20, "grade":"A"}
print(student)
```

```python
## Accessing elements in a dictionary
print(student["name"])
print(student["age"])

## Accessing using get() method
print(student.get("name"))
print(student.get("last_name"))
print(student.get("last_name", "Not Found"))  # Default value if key doesn't exist
```

```python
## Modifying Dictionary Elements
## Dictionary are mutable,so you can add, update or delete elements
print(student)
```

```python
student["age"]=33 # Update age
print(student)
student["course"]="Python" # Add new key-value pair
print(student)
```

```python
del student["grade"] # Delete a key-value pair
print(student)
```

```python
## Dictionary Methods

keys=student.keys() ##get all the keys
print(keys)
values=student.values() ##get all values
print(values)
items=student.items() ##get all key-value pairs
print(items)
```

```python
## shalow copy
student_copy=student
print(student)
print(student_copy)
```

```python
student["name"]="Doe" # Modify original dictionary
print(student)
print(student_copy) # Copy reflects the change
```

```python
student_copy1=student.copy()  # Create a shallow copy
print(student_copy1)
print(student)  # Check if they are the same object
```

```python
student["name"]="jack"
print(student_copy1)  # Original dictionary is modified, but copy remains unchanged
print(student)  # Check if they are the same object
```

```python
## iterating over dictionares 
## You can use loops to iterate over dictionatries, keys,values,or items

for keys in student.keys():
    print(keys)
```

```python
## iterate over values 
for value in student.values():
    print(value)
```

```python
## iterate over items
for key, value in student.items():
    print(f"{key}: {value}")
```

```python
## Nested dictionaries
student={
    "student1": {"name": "John", "age": 20},
    "student2": {"name": "Jane", "age": 22},
}
print(student)
```

```python
## Access nested ditionaries elements
print(student["student1"]["name"])  # Accessing nested dictionary
print(student["student2"]["age"])   # Accessing nested dictionary
```

```python
student.items()
```

```python
## iterating over nested dictionaries 
for student_id,student_info in student.items():
    print(f"{student_id}:{student_info}")
    for key,value in student_info.items():
        print(f"{key}:{value}")
```

```python
## Dictionary comprehension 
squares={x**2 for x in range(1, 11)}
print(squares) 
```

```python
## Conditional dictionaries 
squares={}
for x in range(1,11):
    if x%2==0:
        squares[x]=x**2
print(squares)  # Print the squares of even numbers from 1 to 10
```

```python
## Practical exmaples

## use a dictionary to count the frequency of elemnts in list 

numbers=[1,2,2,3,3,3,4,4,4,4]
frequency={}

for number in numbers:
    if number in frequency:
        frequency[number] += 1
    else:
        frequency[number] = 1
```

```python
print(frequency)
```

```python
## Merge 2 dictonaries into one 

dict1={"a":1, "b":2}
dict2={"c":3, "d":4}
merged_dict={**dict1, **dict2}  # Merging dictionaries
print(merged_dict)  # Print the merged dictionary
```

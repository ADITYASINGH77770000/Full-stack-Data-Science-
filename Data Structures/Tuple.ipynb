```python
## Immutable Nature of Tuples
## Tuples are immutable, meaning their elements cannot be changed after creation.

lst = [1,2,3,4,5]
print(lst)

lst[1]="Aditya"
print(lst)
```

```python
numbers[1]="Aditya" # This will raise an error because tuples are immutable
```

```python
## Tuple Methods
print(numbers.count(2))  # Count occurrences of 2 in the tuple
print(numbers.index(3))  # Find the index of the first occurrence of 3 in
```

```python
## Packing and unpacking tuple
## packing 
packed_tuple=1,"Hello",3.14
print(packed_tuple)
```

```python
## Unapcking a tuple
a,b,c=packed_tuple
print(a)
print(b)
print(c)
```

```python
## Unapcking with *
numbers=(1,2,3,4,5,6)
first,*middle,last=numbers
print(first)
print(middle)
print(last)
```

```python
## Nested Tuple
## Nested list 
lst=[[1,2,3],[4,5,6],[1,"hello",3.14,"c"]]
lst[0][0:2]
```

```python
lst=[[1,2,3,4],[6,7,8,9],(1,"Hello",3.14,"c")]
lst[2][0:3]
```

```python
nested_tuple=((1,2,3),(4,5,6),(1,"hello",3.14,"c"))
nested_tuple[2][0:3]
```

```python
## Iterating over nested tuple 
for sub_tuple in nested_tuple:
    for item in sub_tuple:
        print(item, end=' ')
    print()  # New line after each sub-tuple
```

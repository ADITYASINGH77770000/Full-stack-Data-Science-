```python
## Accessing List Elements
fruits=["apple", "banana", "cherry"]
```

```python
print(fruits[0])  # Accessing the first element
print(fruits[1])  # Accessing the second element
print(fruits[2])  # Accessing the third element
print(fruits[-1]) # Accessing the last element
```

```python
print(fruits[1:])
print(fruits[1:3])
```

```python
fruits[1]="orange"  # Changing the second element
print(fruits)
```

```python
fruits=["apple", "banana", "cherry"]
```

```python
## List methods

fruits.append("orange")  # Adding an element to the end
print(fruits)
```

```python
fruits.insert(1,"watermelon")  # Inserting an element at a specific position
print(fruits)   
```

```python
fruits.remove("banana")  # Removing an element by value
print(fruits)
```

```python
## Remove and return the last elemnts 

last_fruit = fruits.pop()  # Removing the last element
print(last_fruit)  # Output: orange
print(fruits)  # Output: ['apple', 'watermelon', 'cherry']
```

```python
index=fruits.index("apple")  # Finding the index of an element
print(index)  # Output: 0
```

```python
fruits.insert(2,"banana")  # Inserting "banana" at index 2
print(fruits) 
```

```python
fruits.sort()  # Sorting the list
```

```python
fruits
```

```python
fruits.reverse()  # Reversing the list
print(fruits)
```

```python
fruits.clear()  # Clearing the list
print(fruits)  
```

```python
## Slicing List

fruits = ["apple", "banana", "cherry", "date", "elderberry"]
print(fruits[1:3])  # Output: ['banana', 'cherry']
print(fruits[:2])   # Output: ['apple', 'banana']
print(fruits[2:])   # Output: ['cherry', 'date', 'elderberry']
print(fruits[-3:])  # Output: ['cherry', 'date', 'elderberry']
print(fruits[::2])  # Output: ['apple', 'cherry', 'elderberry']
print(fruits[::-1])  # Output: ['elderberry', 'date', 'cherry', 'banana', 'apple']
```

```python
print(fruits[::3])
```

```python
## Iterating over a list 

for number in fruits:
    print(fruits)
```

```python
## Iterating with index
for index,number in enumerate(fruits):
    print(f"Index: {index}, Value: {number}")
```

```python
## List comprehension
lst=[]
for x in range(10):
    lst.append(x**2)    
```

```python
print(lst)
```

```python
## Basic list comprehension
squared_numbers = [x**2 for x in range(10)]
print(squared_numbers)
```

```python
### List comprehension with condition
lst=[]
for x in range(11):
    if x%2==0:
        lst.append(x**2)
```

```python
print(lst)
```

```python
even_numbers=[x**2 for x in range(10) if x%2==0]
print(even_numbers)
```

```python
## Nested List Comprehension

lst1=[1,2,3,4] 
lst2=['a','b','c','d']
pair=[[i,j] for i in lst1 for j in lst2]

print(pair)
```

```python
## List Comprehension with function calls
words = ["hello", "world", "python"]
lengths = [len(word) for word in words]
print(lengths) 
```
